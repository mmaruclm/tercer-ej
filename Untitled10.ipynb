```python
import numpy as np

def cartesian_to_spherical_coords(x, y, z, vx, vy, vz):
    # C√°lculo de las coordenadas esf√©ricas
    r = np.sqrt(x**2 + y**2 + z**2)
    theta = np.arccos(z / r)
    phi = np.arctan2(y, x)
    
    # Matriz de transformaci√≥n
    transformation_matrix = np.array([
        [np.sin(theta) * np.cos(phi), np.sin(theta) * np.sin(phi), np.cos(theta)],
        [np.cos(theta) * np.cos(phi), np.cos(theta) * np.sin(phi), -np.sin(theta)],
        [-np.sin(phi), np.cos(phi), 0]
    ])
    
    # Vector en coordenadas cartesianas
    vector_cartesian = np.array([vx, vy, vz])
    
    # Transformaci√≥n del vector
    vector_spherical = np.dot(transformation_matrix, vector_cartesian)
    
    vr, vtheta, vphi = vector_spherical
    
    return r, theta, phi, vr, vtheta, vphi

# Ejemplo de uso œï y 

x, y, z = 1, 1, 1
vx, vy, vz = 1, 0, 0
r, theta, phi, vr, vtheta, vphi = cartesian_to_spherical_coords(x, y, z, vx, vy, vz)

print(f"Coordenadas esf√©ricas: r={r}, ùúÉ={theta}, œï ={phi}")
print(f"Vector en coordenadas esf√©ricas: vr={vr}, vùúÉ={vtheta}, vœï={vphi}")
```

```python
import numpy as np
import matplotlib.pyplot as plt


omega = 1.0  
mu0 = 4 * np.pi * 1e-7  
k = 1.0  #(valor arbitrario)

# componentes del campo en coordenadas esf√©ricas
def E_r(r, theta, phi):
    return omega**2 * mu0 * np.exp(1j * k * r) / r * (1 - np.sin(theta)**2 * np.cos(phi)**2)

def E_theta(r, theta, phi):
    return -omega**2 * mu0 * np.exp(1j * k * r) / r * np.sin(theta)**2 * np.sin(phi) * np.cos(phi)

def E_phi(r, theta, phi):
    return -omega**2 * mu0 * np.exp(1j * k * r) / r * np.sin(theta) * np.cos(theta)


r_values = np.logspace(0, 3, 100)  
theta = np.pi / 4  # 45 grados
phi = np.pi / 4  # 45 grados


E_r_values = np.abs(E_r(r_values, theta, phi))
E_theta_values = np.abs(E_theta(r_values, theta, phi))
E_phi_values = np.abs(E_phi(r_values, theta, phi))

# Graficar los resultados
plt.figure(figsize=(10, 6))
plt.plot(r_values, np.log10(E_r_values), label='$|E_r|$')
plt.plot(r_values, np.log10(E_theta_values), label='$|E_\\theta|$')
plt.plot(r_values, np.log10(E_phi_values), label='$|E_\\phi|$')

plt.xlabel('r')
plt.ylabel('Log(Magnitude of Field Components)')
plt.title('Field Components in Spherical Coordinates')
plt.legend()
plt.grid(True)
plt.xscale('log')
plt.show()
```
